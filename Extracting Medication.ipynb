```python
medicationmention = []
ontConceptArr = {}
for atype in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationMention'):
    medicationmention.append(atype.get('_id')) # get the disease ids
    ontConceptArr[atype.get('_id')] = atype.get('_ref_ontologyConceptArr') # map the disease ids to the oca ids
```

```python
extra_medical_features = []


for atype in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationMention'):
    if atype.get('_id') in medicationmention:
        medical_features = {}
        medical_features['M_id'] = atype.get('_id')
        
        medStatusChange = atype.get('_ref_medicationStatusChange')
        for ch in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationStatusChangeModifier'):
            if medStatusChange == ch.get('_id'):
                medical_features['Status Change'] = ch.get('category')
            
        
        med_freq = atype.get('_ref_medicationFrequency')
        if med_freq != None:
            for r in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationFrequencyModifier'):
                if med_freq == r.get('_id'):
                    a = r.get('_ref_normalizedForm')
                    for p in e.findall('org.apache.ctakes.typesystem.type.refsem.MedicationFrequency'):
                        if a == p.get('_id'):
                            medical_features['med_freq'] = p.get('unit')
                    

        med_route = atype.get('_ref_medicationRoute')
        if med_route != None:
            for r in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationRouteModifier'):
                if med_route == r.get('_id'):
                    a = r.get('_ref_normalizedForm')
                    for t in e.findall('org.apache.ctakes.typesystem.type.refsem.MedicationRoute'):
                        if a == t.get('_id'):
                            medical_features['Medicine Intake Route'] = t.get('value')
                    




        med_strength = atype.get('_ref_medicationStrength')
        if med_strength != None:
            for r in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationStrengthModifier'):
                if med_strength == r.get('_id'):
                    medical_features['Medicine Strength'] = r.get('category')
            


        med_duration = atype.get('_ref_medicationDuration')
        if med_duration !=None:
            for t in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationDurationModifier'):
                if med_duration == t.get('_id'):
                    medical_features['Medicine Duration'] = t.get('category')


        extra_medical_features.append(medical_features)
        

        
#extra_medical_features

```

```python
oca_umls_children = []
for i in ontConceptArr.keys():
    the_oca = ontConceptArr[i]
    for atype in e.findall('uima.cas.FSArray'):
        if the_oca == atype.get('_id'):
            the_kids = []
            for child in atype.getchildren():
                the_kids.append(child.text)
            oca_umls_children.append({the_oca : the_kids})
```

```python
#declaring all the variables
m_id = 0
oca_id = 0
umls_id = 0
medication = 0
uncertainty = 0
polarity = 0
confidence = 0
discoveryTechnique = 0
conditional = 0
generic = 0
subject = 0
historyOf = 0
StatusChange = 0
MedicationFrequency = 0
MedicationRoute = 0
MedicationStrength = 0
MedicationDuration = 0
codingScheme = 0
code = 0
oid = 0
cui = 0
tui = 0
```

```python
import pandas as pd

df_medicine= pd.DataFrame(columns = ['d_id','oca_id','umls_id','medication','uncertainty','polarity',
                                    'confidence','discoveryTechnique','conditional','generic','subject','historyOf',
                                   'StatusChange','MedicationFrequency','MedicationRoute','MedicationStrength',
                                   'MedicationDuration','codingScheme','code','oid','cui','tui'])

count = 0
listed = ['StatusChange','MedicationFrequency','MedicationRoute','MedicationStrength','MedicationDuration']
listed1 = []
for i in ontConceptArr:
    m_id = i # giving the medicine id
    oca_id = ontConceptArr[i] # giving the oca corresponding to the medicine id
    
    k = extra_medical_features[count]
    
    if 'Status Change' in k.keys():
        StatusChange = k['Status Change']
    else:
        StatusChange = 'Not mentioned'
        
    if 'med_freq' in k.keys():
        MedicationFrequency = k['med_freq']
    else:
        MedicationFrequency = 'Not Mentioned'
        
    if 'Medicine Intake Route' in k.keys():
        MedicationRoute = k['Medicine Intake Route']
    else:
        MedicationRoute = 'Not Mentioned'
    if 'Medicine Strength' in k.keys():
        MedicationStrength = k['Medicine Strength']
    else:
        MedicationStrength = 'Not Mentioned'
    if 'Medicine Duration' in k.keys():
        MedicationDuration = k['Medicine Duration']
    else:
        MedicationDuration = 'Not Mentioned'
            
    for j in range(len(oca_umls_children)): # go through each element at(0,1,2,..) in 
        if oca_id in oca_umls_children[j]:
            umls_id = oca_umls_children[j][oca_id][0]
            
    for atype in e.findall('org.apache.ctakes.typesystem.type.refsem.UmlsConcept'):
        if umls_id == atype.get('_id'):
            medication = atype.get('preferredText')
            codingScheme = atype.get('codingScheme')
            oid = atype.get('oid')
            cui = atype.get('cui')
            code = atype.get('code')
            tui = atype.get('tui')
            
    for atype in e.findall('org.apache.ctakes.typesystem.type.textsem.MedicationMention'):
        if oca_id == atype.get('_ref_ontologyConceptArr'):
            uncertainty = atype.get('uncertainty')
            polarity = atype.get('polarity')
            confidence = atype.get('confidence')
            discoveryTechnique = atype.get('discoveryTechnique')
            conditional = atype.get('conditional')
            generic = atype.get('generic')
            subject = atype.get('subject')
            historyOf = atype.get('historyOf')
            
    df_medicine = df_medicine.append({'d_id':d_id,'oca_id':oca_id,'umls_id':umls_id,'medication':medication,
                                      'uncertainty':uncertainty,
                                    'polarity':polarity,'confidence':confidence,'discoveryTechnique':discoveryTechnique,
                                    'conditional':conditional,'generic':generic,'subject':subject,'historyOf':historyOf,
                                   'StatusChange' : StatusChange,'MedicationFrequency':MedicationFrequency,
                                   'MedicationRoute':MedicationRoute,'MedicationStrength' : MedicationStrength,
                                   'MedicationDuration' : MedicationDuration,'codingScheme':codingScheme,
                                      'code':code,'oid':oid,'cui':cui,'tui':tui},
                                   ignore_index = True)
    count = count + 1
```
